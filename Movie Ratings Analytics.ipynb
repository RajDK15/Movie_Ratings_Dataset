```python
## back to scatterplots:
r_1 = sns.FacetGrid(movies, row = 'Genre', col = 'Year of release', hue = 'Genre')
kws = dict(s= 50, linewidth = 0.5, edgecolor = 'black')
r_1 = r_1.map(plt.scatter, "Rotten Tomatoes Ratings %", "Audience Ratings %", **kws)
```

```python
## Controlling Axes and adding Diagonals
r_1 = sns.FacetGrid(movies, row='Genre', col='Year of release', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='black')
r_1 = r_1.map(sns.scatterplot, "Rotten Tomatoes Ratings %", "Audience Ratings %", **kws)
r_1.set(xlim=(0, 100), ylim=(0, 100))
for ax in r_1.axes.flat:
    ax.plot((0,100), (0,100), c= 'gray', ls = '--')
r_1.add_legend()    
plt.show()
```

```python
## Building dashboards
sns.set_style("darkgrid")
f, axes = plt.subplots(2, 2, figsize=(15, 15))
k_1 = sns.kdeplot(x=movies['Budget (million $)'], y=movies['Audience Ratings %'], ax=axes[0,0])
k_2 = sns.kdeplot(x=movies['Budget (million $)'], y=movies['Rotten Tomatoes Ratings %'], ax=axes[0,1])
k_1.set(xlim=(-20, 160))
k_2.set(xlim=(-20, 160))
w = sns.violinplot(data=movies[movies['Genre'] == 'Drama'], x='Year of release', y='Rotten Tomatoes Ratings %', ax = axes[1,0])
k_4 = sns.kdeplot(x = movies['Rotten Tomatoes Ratings %'], y = movies['Audience Ratings %'],
                shade = True, shade_lowest = False, cmap = 'Reds', ax = axes[1,1])
# to add hist plot we have to use the below code because hist chart is not a seaborn plot and has a different approach
# axes[1,1].hist(movies['Rotten Tomatoes Ratings %'], bins = 15) 
plt.show()
```

```python
## Styling dashboards
sns.set_style("white", {'axes.facecolor':'black'})
f, axes = plt.subplots(2, 2, figsize=(15, 15))

k_1 = sns.kdeplot(x=movies['Budget (million $)'], y=movies['Audience Ratings %'], ax=axes[0,0] \
                 ,shade = True, shade_lowest = True, cmap = 'inferno')
## for the outline of the plot
k_1 = sns.kdeplot(x=movies['Budget (million $)'], y=movies['Audience Ratings %'], ax=axes[0,0] \
                 ,cmap = 'Blues')

k_2 = sns.kdeplot(x=movies['Budget (million $)'], y=movies['Rotten Tomatoes Ratings %'], ax=axes[0,1],\
                  shade = True, shade_lowest = True, cmap = 'cool')
k_2 = sns.kdeplot(x=movies['Budget (million $)'], y=movies['Rotten Tomatoes Ratings %'], ax=axes[0,1],\
                  cmap = 'summer')

w = sns.violinplot(data=movies[movies['Genre'] == 'Drama'], x='Year of release', y='Rotten Tomatoes Ratings %', ax = axes[1,0],
                  palette = 'YlOrRd')

k_4 = sns.kdeplot(x = movies['Rotten Tomatoes Ratings %'], y = movies['Audience Ratings %'],
                shade = True, shade_lowest = False, cmap = 'Blues', ax = axes[1,1])
k_4 = sns.kdeplot(x = movies['Rotten Tomatoes Ratings %'], y = movies['Audience Ratings %'],
                cmap = 'gist_gray', ax = axes[1,1])

# to add hist plot we have to use the below code because hist chart is not a seaborn plot and has a different approach
# axes[1,1].hist(movies['Rotten Tomatoes Ratings %'], bins = 15) 
k_1.set(xlim=(-20, 160))
k_2.set(xlim=(-20, 160))
plt.show()
```
